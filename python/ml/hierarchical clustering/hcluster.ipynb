```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("/Users/lambardaar/Documents/study_hub/python/ml/cluster.csv")
df.head()

```

```python
data=df[["AGE","HOUR","INCOME"]]
data


from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)



import scipy.cluster.hierarchy as shc
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 7))
shc.dendrogram(shc.linkage(data_scaled, method='ward'))
plt.title("Dendrogram")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()





from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(
    n_clusters=2,
    linkage='ward'
)
labels_ = cluster.fit_predict(data_scaled)




plt.figure(figsize=(6, 5))
plt.scatter(
    data_scaled[:, 0],  # AGE
    data_scaled[:, 1],  # HOUR
    c=labels_,
    cmap='rainbow'
)
plt.xlabel("AGE (scaled)")
plt.ylabel("HOUR (scaled)")
plt.title("Hierarchical Clustering Scatter Plot")
plt.show()
```
