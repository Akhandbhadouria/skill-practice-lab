```python
import pandas as pd
df=pd.read_csv("/Users/lambardaar/Documents/study_hub/python/ml/class_practice/dataset/train.csv")
df.head(3)
df.shape
```

```python
import matplotlib.pyplot as plt 
plt.imshow(df.iloc[32473,1:].values.reshape(28,28))
```

```python
X=df.iloc[:,1:].values
y=df.iloc[:,0].values

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)

X_train.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn=KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2)
knn.fit(X_train,y_train)

y_pred=knn.predict(X_test)

from sklearn.metrics import accuracy_score

accuracy_score(y_test,y_pred)

```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

from sklearn.decomposition import PCA
pca = PCA(n_components=200)
X_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
```

```python
X_pca.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn=KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2)
knn.fit(X_train,y_train)
y_pred=knn.predict(X_test)

from sklearn.metrics import accuracy_score

accuracy_score(y_test,y_pred)

```

```python
#apply for loop to change number of components for getting good accuracy    
from sklearn.decomposition import PCA
pca=PCA()
pca.fit(X_train)
X_train_pca=pca.transform(X_train)
X_test_pca=pca.transform(X_test)
for i in range(1,10):
    knn=KNeighborsClassifier(n_neighbors=i,metric='minkowski',p=2)
    knn.fit(X_train_pca,y_train)
    y_pred=knn.predict(X_test_pca)
    print(accuracy_score(y_test,y_pred))
```
